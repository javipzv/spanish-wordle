```python
path_file = "static/0_palabras_todas_no_conjugaciones.txt"
```

```python
with open(path_file, encoding="utf-8") as file:
    lines = file.readlines()

print(lines)
```

```python
voc = [x.replace("\n", "") for x in lines]
voc = list(filter(lambda x: len(x) == 5, voc))
voc = list(filter(lambda x: not "á" in x and not "é" in x and not "í" in x and not "ó" in x and not "ú" in x, voc))
voc
```

```python
with open("static/vocabulario_5_letras_sin_conjugaciones.txt", "w", encoding="utf-8") as file:
    for word in voc:
        file.write(word + "\n")
```
